```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov5s.pt')
```

```python
# Example: Modify dataset.yaml after copying it
import yaml

# 1. Copy the original file to working directory
!cp /kaggle/input/traffic-sign/data/data.yaml /kaggle/working/

# 2. Load the YAML file
with open('/kaggle/working/data.yaml', 'r') as f:
    data = yaml.safe_load(f)

# 3. Modify the content
data['train'] = '/kaggle/input/traffic-sign/data/images/train'  # Update paths
data['val'] = '/kaggle/input/traffic-sign/data/images/val'
data['test'] = '/kaggle/input/traffic-sign/data/images/test'

# 4. Save the modified file
with open('/kaggle/working/data.yaml', 'w') as f:
    yaml.dump(data, f)
```

```python
results = model.train(
    data='/kaggle/working/data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,  # use GPU
    workers=2,
    project='yolov5_kaggle',
    name='exp'
)
```

```python
model = YOLO('/kaggle/working/yolov5_kaggle/exp5/weights/best.pt')

# Validate
results = model.val(
    data='/kaggle/working/data.yaml',
    batch=16,
    imgsz=640,
    conf=0.25,  # confidence threshold
    iou=0.45,   # IoU threshold
    device=0    # GPU
)

# Print metrics
print(f"mAP@0.5: {results.box.map}")
print(f"mAP@0.5:0.95: {results.box.map75}")
```

```python
results = model.predict(
    source='/kaggle/input/traffic-sign/data/images/test',  # or folder path
    save=True,            # save results
    save_txt=True,        # save results as .txt
    save_conf=True,       # save confidence scores
    save_crop=False,      # save cropped predictions
    device=0              # GPU
)
```

```python
results = model.predict(
    source='/kaggle/input/traffic-sign/data/video.mp4',
    save=True,
    device=0
)
```
